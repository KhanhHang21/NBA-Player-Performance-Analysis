## Hypothesis Testing

### p-Value calculation

### Ran regressional model (above)
#### 1) Gives two key things for each independent variable:

1. **coefficient** (how much PPG changes per 1-unit increase)

2. **standard error** (how "noisy" or uncertain the estimate is)

#### 2): For EACH metric (ORtg, DRtg, RPG, APG, MPG):

Uses this formula:

**t-statistic = (coefficient) ÷ (standard error)**

a. Bigger t-statistic → More likely the relationship is real

b. Smaller t-statistic → More likely it's random chance

#### 3) Calculates the p-value: (Use scipy.stats.t to compute it- BELOW)

The p-value = probability that the observed relationship happened by random chance:

• Small p-value (below 0.05) → Statistically significant

• Large p-value (above 0.05) → Not statistically significant

```python
from scipy import stats
```

**(One Metric Example: Offensive Rating)** → p-Value

```python
# Example: ORtg
t_value = 1.599
df = 207  # degrees of freedom from model (observations - predictors - 1)

# p-value
p_value = 2 * stats.t.sf(abs(t_value), df)
print(p_value)  
```

**(ALL Metrics: Offensive Rating)** → p-Values

```python
# Degrees of freedom from your model (observations - predictors - 1)
df = 207

# t-statistics from regression output
t_values = {
    'ORtg': 1.599,
    'DRtg': -2.407,
    'RPG': 3.748,
    'APG': 8.940,
    'MPG': 8.138
}

# Calculate & print p-values for ALL metrics
for metric, t in t_values.items():
    p_value = 2 * stats.t.sf(abs(t), df)  # Two-tailed test
    print(f"{metric}: p-value ≈ {p_value:.3f}")
```

```python
# Load dataset
df = pd.read_csv("NBA Stats 202324 All Stats.csv")
```

```python
#Calculate top 5 composite efficiency scores & players

# Filter who play > 35 min. per game
filtered_df = df[df["MPG"] > 35]

# Select specific columns
metrics = [
    "NAME", "ORtg", "DRtg", "eFG%", "TS%", "BPG", "SPG", "MPG", "PPG", "RPG", "APG"
]
selected_df = filtered_df[metrics].copy()

# Normalize each metric using min-max scaling
for col in metrics[1:]:  # skip "NAME"
    min_val = selected_df[col].min()
    max_val = selected_df[col].max()
    selected_df[col] = (selected_df[col] - min_val) / (max_val - min_val)

# Invert: lower values → indicate better performance
selected_df["DRtg"] = 1 - selected_df["DRtg"]

# Calculate composite efficiency score using all metrics
selected_df["Composite_Efficiency_Score"] = (
    selected_df["ORtg"]
    + selected_df["DRtg"]
    + selected_df["eFG%"]
    + selected_df["TS%"]
    + selected_df["BPG"]
    + selected_df["SPG"]
    + selected_df["MPG"]
    + selected_df["PPG"]
    + selected_df["RPG"]
    + selected_df["APG"]
)

# Show top performers in order
top_players = selected_df.sort_values(by="Composite_Efficiency_Score", ascending=False)
print(top_players[["NAME", "Composite_Efficiency_Score"]].head(5))
```
